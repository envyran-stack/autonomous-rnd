## 2) 다중 선형 회귀

> **다중선형회귀**는 여러 입력변수를 동시에 고려하여 출력변수를 예측하는 모델임
> - 회귀식: $\hat{Y} = \hat{\beta}_0 + \hat{\beta}_1 X_1 + \hat{\beta}_2 X_2 + \cdots + \hat{\beta}_p X_p$
> - **최소자승법(OLS)** 으로 잔차의 제곱합(SSE)을 최소화하는 회귀 계수를 추정함
> - 회귀 계수 $\hat{\beta}_i$는 다른 변수가 고정된 상태에서 $X_i$가 1단위 증가할 때 $Y$의 변화량을 의미함

```python
# 회귀 모형 성능 지표 산출 함수
def get_regscore(true, pred):
    print("MSE       : %.3f" % (mean_squared_error(true, pred)))
    print("RMSE      : %.3f" % (np.sqrt(mean_squared_error(true, pred))))
    print("MAE       : %.3f" % (mean_absolute_error(true, pred)))
    print("R-squared : %.3f" % (r2_score(true, pred)))
```

### 2-1) 모델 학습

```python
LR_model = LinearRegression()
LR_model.fit(X_train, Y_train)
```

### 2-2) 회귀 계수 확인

> 회귀 계수의 부호와 크기를 통해 각 입력변수가 출력변수에 미치는 **영향의 방향과 크기**를 파악할 수 있음

```python
print("절편 (beta0) :", LR_model.intercept_)
print("회귀 계수    :", LR_model.coef_)
```

```python
Coef_LR = pd.DataFrame({"Variable": X_train.columns, "Coefficient": LR_model.coef_})
Coef_LR
```

### 2-3) 성능 평가

> **R-squared (결정계수)** 는 모델이 데이터를 얼마나 잘 설명하는지를 나타내는 지표임
> - $R^2 = 1 - \frac{SSE}{SST}$ 이며, 1에 가까울수록 모델의 설명력이 높음
> - 학습 데이터에서 $R^2$의 최솟값은 0이지만, 테스트 데이터에서는 **음수**가 나올 수 있음

```python
Y_pred_LR = LR_model.predict(X_test)
get_regscore(Y_test, Y_pred_LR)
```

### 2-4) 예측 시각화

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].scatter(Y_train, LR_model.predict(X_train), alpha=0.5)
axes[0].plot([Y_train.min(), Y_train.max()], [Y_train.min(), Y_train.max()], "r--")
axes[0].set_xlabel("Real y")
axes[0].set_ylabel("Predicted y")
axes[0].set_title("Train Data")

axes[1].scatter(Y_test, Y_pred_LR, alpha=0.5)
axes[1].plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], "r--")
axes[1].set_xlabel("Real y")
axes[1].set_ylabel("Predicted y")
axes[1].set_title("Test Data")

plt.tight_layout()
plt.show()
```

## 3) Ridge

> **Ridge 회귀**는 OLS의 손실 함수에 **회귀 계수의 제곱합(L2 규제)** 을 추가하여 계수의 크기를 줄임
>
> $$Minimize \sum(y_i - \hat{y}_i)^2 + \alpha \sum \hat{\beta}_j^2$$
>
> - 모든 변수를 유지하되, 계수의 크기를 **전반적으로 축소**함
> - alpha(λ)가 클수록 규제가 강해져 계수가 0에 가까워짐
> - 입력변수들이 비슷한 수준으로 출력변수에 영향을 미치는 경우에 적합한 모델임

### 3-1) 모델 학습 및 성능 평가 (alpha=0.1)

```python
Ridge_model_1 = Ridge(alpha=0.1)
Ridge_model_1.fit(X_train, Y_train)
Ridge_pred_1 = Ridge_model_1.predict(X_test)
get_regscore(Y_test, Ridge_pred_1)
```

### 3-2) 회귀 계수 확인

```python
Coef_Ridge_1 = pd.DataFrame({"Variable": X_train.columns, "Coefficient": Ridge_model_1.coef_})
Coef_Ridge_1
```

### 3-3) alpha 값 변경 (alpha=10)

```python
Ridge_model_2 = Ridge(alpha=10)
Ridge_model_2.fit(X_train, Y_train)
Ridge_pred_2 = Ridge_model_2.predict(X_test)
get_regscore(Y_test, Ridge_pred_2)
```

```python
Coef_Ridge_2 = pd.DataFrame({"Variable": X_train.columns, "Coefficient": Ridge_model_2.coef_})
Coef_Ridge_2
```

## 4) Lasso

> **Lasso 회귀**는 OLS의 손실 함수에 **회귀 계수의 절대값 합(L1 규제)** 을 추가함
>
> $$Minimize \sum(y_i - \hat{y}_i)^2 + \alpha \sum |\hat{\beta}_j|$$
>
> - Ridge와 달리 불필요한 변수의 계수를 **완전히 0으로** 만들어 변수를 제거함
> - alpha(λ)가 클수록 더 많은 변수가 제거됨
> - 출력변수에 미치는 입력변수의 영향력 편차가 큰 경우에 적합한 모델임

### 4-1) 모델 학습 및 성능 평가 (alpha=0.1)

```python
Lasso_model_1 = Lasso(alpha=0.1)
Lasso_model_1.fit(X_train, Y_train)
Lasso_pred_1 = Lasso_model_1.predict(X_test)
get_regscore(Y_test, Lasso_pred_1)
```

### 4-2) 회귀 계수 확인

```python
Coef_Lasso_1 = pd.DataFrame({"Variable": X_train.columns, "Coefficient": Lasso_model_1.coef_})
Coef_Lasso_1
```

```python
# 계수가 0인 변수 확인 (제거된 변수)
Coef_Lasso_1.loc[Coef_Lasso_1["Coefficient"] == 0]
```

### 4-3) alpha 값 변경 (alpha=10)

```python
Lasso_model_2 = Lasso(alpha=10)
Lasso_model_2.fit(X_train, Y_train)
Lasso_pred_2 = Lasso_model_2.predict(X_test)
get_regscore(Y_test, Lasso_pred_2)
```

### 4-4) 변수 제거 효과 확인

> alpha 값이 커질수록 Lasso는 더 많은 변수의 계수를 0으로 만들어 **자동으로 변수 선택**을 수행합니다.

```python
Coef_Lasso_2 = pd.DataFrame({"Variable": X_train.columns, "Coefficient": Lasso_model_2.coef_})
Coef_Lasso_2
```

```python
# alpha=10에서 제거된 변수
print("제거된 변수:")
print(Coef_Lasso_2.loc[Coef_Lasso_2["Coefficient"] == 0])
print()
print("선택된 변수:")
print(Coef_Lasso_2.loc[Coef_Lasso_2["Coefficient"] != 0])
```

**[과제 1]** 지수는 콘크리트 데이터셋을 처음 접했을 때, 각 변수를 개별적으로 봤을 때(단일 변수 관점), 두 변수씩 짝지어 봤을 때(두 변수 간 관점), 그리고 여러 변수를 한꺼번에 놓고 봤을 때(다변수 관점) 각각 어떤 특성과 관계가 드러나는지 단계적으로 파악하고 싶습니다. 특히 각 설명변수가 압축강도(Y)와 어떤 관계를 갖는지가 가장 궁금합니다. 이 세 가지 관점 각각에 적용할 수 있는 탐색적 데이터 분석(EDA) 방법들을 AI와 상의해서 나열하고, 직접 코드로 탐색해 본 뒤 그 결과가 의미하는 바를 분석해 보세요.

```python
# [과제 1] EDA: 단일 / 두 변수 / 다변수 관점
# ※ 실행 전: 상단 셀(데이터 로드, get_regscore, train_test_split) 먼저 실행

target = "Concrete_compressive_strength"
features = [col for col in Concrete_data.columns if col != target]

# 1) 단일 변수 관점
print("=== 1) 단일 변수 vs Y 상관계수 ===")
single_corr = Concrete_data[features + [target]].corr()[target].drop(target).sort_values(ascending=False)
print(single_corr)

fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()
for i, col in enumerate(features):
    sns.scatterplot(x=Concrete_data[col], y=Concrete_data[target], alpha=0.5, ax=axes[i])
    axes[i].set_title(f"{col} vs {target}\n(corr={single_corr[col]:.3f})")
plt.tight_layout()
plt.show()

# 2) 두 변수 간 관점
print("\n=== 2) 두 변수 간 상관계수 (상위 절댓값) ===")
pair_corr = Concrete_data[features].corr()
pair_corr_abs = pair_corr.where(np.triu(np.ones(pair_corr.shape), k=1).astype(bool)).stack()
print(pair_corr_abs.reindex(pair_corr_abs.abs().sort_values(ascending=False).index).head(6))

key_pairs = [("Water", "Superplasticizer"), ("Cement", "Water"), ("Age", "Cement")]
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, (x_col, y_col) in zip(axes, key_pairs):
    sns.scatterplot(
        x=Concrete_data[x_col], y=Concrete_data[y_col],
        hue=Concrete_data[target], palette="viridis", alpha=0.7, ax=ax
    )
    ax.set_title(f"{x_col} vs {y_col} (색=강도)")
plt.tight_layout()
plt.show()

# 3) 다변수 관점
print("\n=== 3) 다변수 관점 ===")
plt.figure(figsize=(10, 8))
sns.heatmap(Concrete_data.corr(numeric_only=True), annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1, center=0)
plt.title("다변수 상관행렬")
plt.tight_layout()
plt.show()

sns.pairplot(Concrete_data[["Cement", "Water", "Age", target]], diag_kind="hist", corner=True, plot_kws={"alpha": 0.4})
plt.show()

print("\n=== EDA 해석 요약 ===")
print(f"- Y와 가장 강한 양의 상관: {single_corr.idxmax()} ({single_corr.max():.3f})")
print(f"- Y와 가장 강한 음의 상관: {single_corr.idxmin()} ({single_corr.min():.3f})")
print("- Water-Superplasticizer: 음의 상관 (감수제↑ → 물↓)")
print("- 다변수 관점에서 다중공선성도 함께 확인 필요")
```

```python
# [과제 2] 파생변수 생성 및 선형회귀 성능 비교

df_feat = Concrete_data.copy()
df_feat["Binder"] = df_feat["Cement"] + df_feat["Blast_Furnace_Slag"] + df_feat["Fly_Ash"]
df_feat["WC_ratio"] = df_feat["Water"] / df_feat["Cement"].replace(0, np.nan)
df_feat["Effective_Water"] = df_feat["Water"] - 10 * df_feat["Superplasticizer"]
df_feat["effective_w_b"] = df_feat["Effective_Water"] / df_feat["Binder"].replace(0, np.nan)

Y_base = df_feat[target]
X_base = df_feat[features]
X_enhanced = df_feat.drop([target], axis=1)

Xb_train, Xb_test, Yb_train, Yb_test = train_test_split(X_base, Y_base, test_size=0.3, random_state=0)
Xe_train, Xe_test, Ye_train, Ye_test = train_test_split(X_enhanced, Y_base, test_size=0.3, random_state=0)

LR_base = LinearRegression().fit(Xb_train, Yb_train)
LR_feat = LinearRegression().fit(Xe_train, Ye_train)

print("=== Baseline (원본 변수) ===")
get_regscore(Yb_test, LR_base.predict(Xb_test))

print("\n=== 파생변수 추가 ===")
get_regscore(Ye_test, LR_feat.predict(Xe_test))
```

### 6-3) Ridge/Lasso 모델링 및 고도화 심화

**[과제 3]** 지수는 Ridge와 Lasso의 alpha를 0.1, 10 두 값만 비교했는데, 최적의 alpha가 그 사이 어딘가에 있을 수도 있다고 생각합니다. 직접 여러 alpha 값을 하나씩 대입해보는 것 말고 더 체계적으로 최적 alpha를 찾는 방법을 AI와 상의해서 정하고, 실제로 최적 alpha를 탐색해 보세요.

```python
# [과제 3] GridSearchCV로 최적 alpha 탐색
from sklearn.model_selection import GridSearchCV

alpha_grid = np.logspace(-2, 2, 30)

ridge_search = GridSearchCV(Ridge(), param_grid={"alpha": alpha_grid}, cv=5, scoring="neg_mean_squared_error")
ridge_search.fit(X_train, Y_train)

lasso_search = GridSearchCV(Lasso(max_iter=10000), param_grid={"alpha": alpha_grid}, cv=5, scoring="neg_mean_squared_error")
lasso_search.fit(X_train, Y_train)

print("=== GridSearchCV 결과 ===")
print(f"Ridge best alpha : {ridge_search.best_params_['alpha']:.4f}")
print(f"Lasso best alpha : {lasso_search.best_params_['alpha']:.4f}")

ridge_cv_mse = -ridge_search.cv_results_["mean_test_score"]
lasso_cv_mse = -lasso_search.cv_results_["mean_test_score"]

plt.figure(figsize=(10, 5))
plt.semilogx(alpha_grid, ridge_cv_mse, marker="o", label="Ridge CV MSE")
plt.semilogx(alpha_grid, lasso_cv_mse, marker="o", label="Lasso CV MSE")
plt.axvline(ridge_search.best_params_["alpha"], color="C0", ls="--", alpha=0.7)
plt.axvline(lasso_search.best_params_["alpha"], color="C1", ls="--", alpha=0.7)
plt.xlabel("alpha"); plt.ylabel("CV MSE"); plt.title("alpha에 따른 교차검증 MSE")
plt.legend(); plt.tight_layout(); plt.show()

ridge_best = ridge_search.best_estimator_
lasso_best = lasso_search.best_estimator_

print("\n=== 최적 alpha 모델 테스트 성능 ===")
print("[Ridge]"); get_regscore(Y_test, ridge_best.predict(X_test))
print("\n[Lasso]"); get_regscore(Y_test, lasso_best.predict(X_test))
```

**[과제 4]** 지수는 Ridge(L2)와 Lasso(L1)의 장점을 함께 활용할 수 있는 방법이 없는지 궁금해졌습니다. AI와 함께 적절한 모델을 찾아보고, 해당 모델의 특징과 활용 상황을 정리한 뒤 직접 적용하여 Ridge, Lasso와 성능 및 회귀 계수를 비교해 보세요.

```python
# [과제 4] Elastic Net (L1 + L2 규제 결합)
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV

enet_search = GridSearchCV(
    ElasticNet(max_iter=10000),
    param_grid={"alpha": np.logspace(-2, 2, 20), "l1_ratio": [0.1, 0.3, 0.5, 0.7, 0.9]},
    cv=5, scoring="neg_mean_squared_error"
)
enet_search.fit(X_train, Y_train)
enet_best = enet_search.best_estimator_

print("=== Elastic Net 최적 하이퍼파라미터 ===")
print(enet_search.best_params_)

compare_rows = []
for name, model in {"Ridge": ridge_best, "Lasso": lasso_best, "ElasticNet": enet_best}.items():
    pred = model.predict(X_test)
    compare_rows.append({"Model": name, "RMSE": np.sqrt(mean_squared_error(Y_test, pred)),
                         "MAE": mean_absolute_error(Y_test, pred), "R2": r2_score(Y_test, pred)})
print("\n=== 성능 비교 ===")
print(pd.DataFrame(compare_rows))

coef_compare = pd.DataFrame({"Variable": X_train.columns, "Ridge": ridge_best.coef_,
                              "Lasso": lasso_best.coef_, "ElasticNet": enet_best.coef_})
print("\n=== 회귀 계수 비교 ===")
print(coef_compare)

x_pos = np.arange(len(X_train.columns)); width = 0.25
plt.figure(figsize=(12, 6))
plt.bar(x_pos - width, coef_compare["Ridge"], width, label="Ridge")
plt.bar(x_pos, coef_compare["Lasso"], width, label="Lasso")
plt.bar(x_pos + width, coef_compare["ElasticNet"], width, label="ElasticNet")
plt.xticks(x_pos, X_train.columns, rotation=45)
plt.ylabel("Coefficient"); plt.title("Ridge / Lasso / ElasticNet 회귀 계수 비교")
plt.legend(); plt.tight_layout(); plt.show()
```

**[과제 5]** 지수는 분석 과정에서 양생 기간(Age)이 길어질수록 강도가 증가하지만, 그 증가 폭은 점점 줄어드는 비선형 관계일 수 있다는 생각이 들었습니다. 선형 회귀 모델이 이런 비선형 관계를 잘 잡아내지 못할 수 있다는 점을 AI와 논의하고, 비선형성을 반영할 수 있는 방법을 적용해 성능이 개선되는지 확인해 보세요.

```python
# [과제 5] Age 비선형성 반영
from sklearn.preprocessing import PolynomialFeatures

X_nonlinear = X_train.copy()
X_nonlinear_test = X_test.copy()
for df_ in [X_nonlinear, X_nonlinear_test]:
    df_["Age_log"] = np.log1p(df_["Age"])
    df_["Age_sqrt"] = np.sqrt(df_["Age"])
    df_["Age_sq"] = df_["Age"] ** 2

poly = PolynomialFeatures(degree=2, include_bias=False)
age_poly_train = poly.fit_transform(X_train[["Age"]])
age_poly_test = poly.transform(X_test[["Age"]])
age_poly_cols = [f"Age_poly_{c}" for c in poly.get_feature_names_out(["Age"])]

X_poly = pd.concat([X_train.reset_index(drop=True), pd.DataFrame(age_poly_train, columns=age_poly_cols)], axis=1)
X_poly_test = pd.concat([X_test.reset_index(drop=True), pd.DataFrame(age_poly_test, columns=age_poly_cols)], axis=1)

LR_linear = LinearRegression().fit(X_train, Y_train)
LR_nonlinear = LinearRegression().fit(X_nonlinear, Y_train)
LR_poly = LinearRegression().fit(X_poly, Y_train)

print("=== Baseline (Age 선형) ===")
get_regscore(Y_test, LR_linear.predict(X_test))
print("\n=== Age 비선형 변환(log/sqrt/sq) 추가 ===")
get_regscore(Y_test, LR_nonlinear.predict(X_nonlinear_test))
print("\n=== Age 2차 다항항 추가 ===")
get_regscore(Y_test, LR_poly.predict(X_poly_test))

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.scatterplot(x=Concrete_data["Age"], y=Concrete_data[target], alpha=0.5)
plt.title("Age vs Strength (원본)")
plt.subplot(1, 2, 2)
sns.regplot(x=Concrete_data["Age"], y=Concrete_data[target], scatter_kws={"alpha": 0.3}, order=2)
plt.title("Age vs Strength (2차 추세선)")
plt.tight_layout(); plt.show()
```
