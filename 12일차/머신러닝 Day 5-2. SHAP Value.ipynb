```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, accuracy_score, f1_score
from sklearn.metrics import (
    r2_score, mean_squared_error, mean_absolute_error,
    accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
)

import shap
```

## 1) SHAP - 회귀 모델

```python
# 회귀 성능 평가 함수
def get_regscore(true, pred):
    print("MSE       : %.3f" % mean_squared_error(true, pred))
    print("RMSE      : %.3f" % np.sqrt(mean_squared_error(true, pred)))
    print("MAE       : %.3f" % mean_absolute_error(true, pred))
    print("R-squared : %.3f" % r2_score(true, pred))
```

### 1-1) 데이터 및 모델 준비

```python
Regression = pd.read_csv("dataset/day5-2_reg_data.csv")
Regression
```

```python
X_reg = Regression.drop(["y"], axis=1) # 입력변수
Y_reg = Regression["y"] # 출력변수

X_train_reg, X_test_reg, Y_train_reg, Y_test_reg = train_test_split(X_reg, Y_reg, test_size=0.2, random_state=42) # 학습/테스트 데이터 분할
```

```python
rf_reg = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=0) # RandomForest로 모델 학습
rf_reg.fit(X_train_reg, Y_train_reg)
Y_pred_reg = rf_reg.predict(X_test_reg)
get_regscore(Y_test_reg, Y_pred_reg)
```

### 2-2) SHAP 분석

```python
# rf_reg 모델을 해석하기 위한 SHAP 설명자 생성
explainer_reg = shap.TreeExplainer(rf_reg)

# 학습 데이터(X_train_reg)에 대한 SHAP value 계산
# 각 feature가 예측값에 기여하는 정도를 나타냄
shap_values_reg = explainer_reg.shap_values(X_train_reg)
```

#### Summary Plot (Bar)
- 막대가 길수록 예측에 영향을 많이 주는 변수라는 뜻임

```python
shap.summary_plot(shap_values_reg, X_train_reg, plot_type='bar')
```

#### Summary Plot (Scatter)
- 오른쪽에 점이 몰려있으면 예측값을 올리는 변수, 왼쪽에 몰려있으면 예측값을 낮추는 변수임

> - 빨간색: 높은 변수 값
> - 파란색: 낮은 변수 값

```python
shap.summary_plot(shap_values_reg, X_train_reg)
```

#### Force Plot (개별 데이터)
- 데이터 하나를 뽑아서 어떤 변수가 예측값을 올렸고 내렸는지 화살표로 보여주는 그래프임

```python
base_value_reg = explainer_reg.expected_value
print(base_value_reg)
```

```python
# 첫 번째 데이터(index 0)의 예측 결과를 force plot으로 시각화
# expected_value: 전체 데이터의 평균 예측값 (기준선)
# shap_values_reg[0]: 이 데이터의 각 feature별 SHAP value 
# X_train_reg.iloc[0]: 실제 feature 값들 (그래프에 표시용)

shap.force_plot(
    explainer_reg.expected_value,
    shap_values_reg[0],
    X_train_reg.iloc[0],
    matplotlib=True
)
```

#### Dependence Plot
- 변수 값이 변할 때 SHAP 값(영향력)이 같이 어떻게 움직이는지 보여주는 그래프임

```python
# "X10" feature 값에 따라 SHAP value가 어떻게 변하는지 시각화
# x축: X10의 실제 값, y축: X10의 SHAP value (예측에 미친 영향)
# 자동으로 X10과 가장 상호작용이 큰 feature를 색상으로 표시

shap.dependence_plot(
    "X10",
    shap_values_reg,
    X_train_reg
)
```

## 3) SHAP - 분류 모델

```python
# 분류 성능 평가 함수
def get_classscore(real, pred):
    print("Accuracy  : %.3f" % accuracy_score(real, pred))
    print("Precision : %.3f" % precision_score(real, pred))
    print("Recall    : %.3f" % recall_score(real, pred))
    print("F1-score  : %.3f" % f1_score(real, pred))
    print()
    print("혼동행렬")
    print(confusion_matrix(real, pred))
```

### 3-1) 데이터 및 모델 준비

```python
Classification = pd.read_csv("dataset/day5-2_cls_data.csv")
Classification
```

```python
X_cls = Classification.drop(["Y"], axis=1) # 입력변수
Y_cls = Classification["Y"] # 출력변수

X_train_cls, X_test_cls, Y_train_cls, Y_test_cls = train_test_split(
    X_cls, Y_cls, test_size=0.2, random_state=123, stratify=Y_cls)

rf_cls = RandomForestClassifier(n_estimators=100, random_state=0)
rf_cls.fit(X_train_cls, Y_train_cls)
Y_pred_cls = rf_cls.predict(X_test_cls)
get_classscore(Y_test_cls, Y_pred_cls)
```

## 4) Vibe Coding 실습

**[과제 1]** 

지수는 Random Forest의 Feature Importance와 SHAP Value가 모두 변수의 중요도를 설명하는 방법이라는 것을 배웠습니다.

AI와 함께 Random Forest (회귀/분류)의 feature_importances_와 SHAP Summary Plot 결과를 비교해 보세요.

두 방법에서 중요하게 평가된 변수가 얼마나 비슷한지 확인하고 순위가 다른 변수는 왜 차이가 발생했는지 AI와 토의하여 해석해 보세요.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

def compare_rf_shap(model, shap_values, feature_names, top_n=15, title=""):
    """
    RF feature_importances_ vs SHAP mean(|value|) 비교
    """
    # 1) RF 중요도
    rf_imp = pd.Series(model.feature_importances_, index=feature_names, name="RF_importance")

    # 2) SHAP 중요도 (Summary Plot bar와 동일: mean absolute SHAP)
    shap_arr = np.array(shap_values)
    if shap_arr.ndim == 3:  # 분류: (샘플, 변수, 클래스)
        shap_arr = np.abs(shap_arr).mean(axis=2)  # 클래스 평균
    shap_imp = pd.Series(np.abs(shap_arr).mean(axis=0), index=feature_names, name="SHAP_importance")

    # 3) 병합 + 순위
    df = pd.concat([rf_imp, shap_imp], axis=1)
    df["RF_rank"] = df["RF_importance"].rank(ascending=False, method="min")
    df["SHAP_rank"] = df["SHAP_importance"].rank(ascending=False, method="min")
    df["rank_diff"] = (df["RF_rank"] - df["SHAP_rank"]).abs()
    df = df.sort_values("SHAP_importance", ascending=False)

    # 4) 시각화: 나란히 막대 비교 (상위 top_n)
    top_df = df.head(top_n).sort_values("SHAP_importance")

    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    top_df["RF_importance"].plot(kind="barh", ax=axes[0], color="steelblue")
    axes[0].set_title("Random Forest feature_importances_")
    axes[0].invert_yaxis()

    top_df["SHAP_importance"].plot(kind="barh", ax=axes[1], color="coral")
    axes[1].set_title("SHAP mean(|value|)  [Summary Plot bar 기준]")
    axes[1].invert_yaxis()

    if title:
        fig.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()

    # 5) 산점도: 두 중요도의 상관
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.scatter(df["RF_importance"], df["SHAP_importance"], alpha=0.6)
    for feat in df.head(top_n).index:
        ax.annotate(feat, (df.loc[feat, "RF_importance"], df.loc[feat, "SHAP_importance"]),
                    fontsize=8, alpha=0.8)
    ax.set_xlabel("RF feature_importances_")
    ax.set_ylabel("SHAP mean(|value|)")
    ax.set_title(f"{title} - 중요도 산점도")
    plt.tight_layout()
    plt.show()

    return df
```

```python
compare_reg_df = compare_rf_shap(
    model=rf_reg,
    shap_values=shap_values_reg,
    feature_names=X_train_reg.columns,
    top_n=15,
    title="회귀 모델 (RandomForestRegressor)"
)

# SHAP Summary Plot bar와 직접 비교하고 싶다면
shap.summary_plot(shap_values_reg, X_train_reg, plot_type="bar", max_display=15)
```

```python
# 클래스 1 기준 SHAP (노트북과 동일)
compare_cls_df = compare_rf_shap(
    model=rf_cls,
    shap_values=shap_values_cls[:, :, 1],
    feature_names=X_train_cls.columns,
    top_n=15,
    title="분류 모델 - 클래스 1 기준"
)

shap.summary_plot(shap_values_cls[:, :, 1], X_train_cls, plot_type="bar", max_display=15)
```

```python
def compare_top_features(df, top_k=10):
    """두 방법의 상위 변수가 얼마나 겹치는지 비교"""
    rf_top = set(df.nsmallest(top_k, "RF_rank").index)
    shap_top = set(df.nsmallest(top_k, "SHAP_rank").index)

    common = rf_top & shap_top
    only_rf = rf_top - shap_top
    only_shap = shap_top - rf_top

    # Spearman 순위 상관
    corr, pval = spearmanr(df["RF_rank"], df["SHAP_rank"])

    print(f"=== 상위 {top_k}개 변수 비교 ===")
    print(f"공통 변수 ({len(common)}개): {sorted(common)}")
    print(f"RF에만 있는 변수: {sorted(only_rf)}")
    print(f"SHAP에만 있는 변수: {sorted(only_shap)}")
    print(f"겹침 비율 (Jaccard): {len(common) / len(rf_top | shap_top):.2%}")
    print(f"Spearman 순위 상관: {corr:.3f} (p-value: {pval:.4f})")

    # 비교 테이블
    result = df[["RF_importance", "RF_rank", "SHAP_importance", "SHAP_rank", "rank_diff"]].copy()
    result["in_RF_top"] = result.index.isin(rf_top)
    result["in_SHAP_top"] = result.index.isin(shap_top)
    result["in_both_top"] = result["in_RF_top"] & result["in_SHAP_top"]
    return result.sort_values("rank_diff", ascending=False)

# 회귀
reg_compare = compare_top_features(compare_reg_df, top_k=10)
display(reg_compare.head(15))

# 분류
cls_compare = compare_top_features(compare_cls_df, top_k=10)
display(cls_compare.head(15))
```
