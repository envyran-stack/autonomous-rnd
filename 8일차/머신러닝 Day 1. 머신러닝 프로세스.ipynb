### 2-3) 데이터 분할

- **Train:** 모델이 학습하는 데 사용하는 데이터. 모델은 이 데이터로 패턴을 배움
- **Validation:** 학습 과정에서 모델 성능을 점검하고 하이퍼파라미터를 조정하는 데 사용하는 데이터. 모델 선택 및 하이퍼파라미터 최적화의 기준이 되며, 최종 평가에는 쓰지 않음
- **Test:** 학습이 모두 끝난 후, 모델의 최종 성능을 평가하는 데 사용하는 데이터. 학습 및 최적화 과정에서 한 번도 보지 않은 데이터로 일반화 성능을 측정함

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split

# 입력변수와 출력변수로 분할
X_reg = Regression.drop(["y"], axis=1)
Y_reg = Regression["y"]

X_cla = Classification.drop(["Y"], axis=1)
Y_cla = Classification["Y"]
```

#### 2-4-1) 수치형 변수 스케일링

1. **Min-Max Scaling:** 모든 값을 0~1 범위로 변환 (이상치가 적은 데이터에 적합)
2. **Z-score Scaling:** 평균 0, 표준편차 1이 되도록 변환 (정규분포에 가까운 데이터에 적합)

### 3-3) 성능 평가

```python
from sklearn.metrics import (mean_absolute_error, mean_squared_error,
                             accuracy_score, recall_score, precision_score,
                             f1_score, roc_auc_score, confusion_matrix, r2_score)

# [회귀 성능지표] MAE, MSE, MAPE를 계산해 딕셔너리로 반환
def get_reg_score(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)                   # 평균 절대 오차
    mse = mean_squared_error(y_true, y_pred)                    # 평균 제곱 오차
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100    # 평균 절대 백분율 오차(%)
    r2 = r2_score(y_true, y_pred) 
    return {"MAE": mae, "MSE": mse, "MAPE(%)": mape, "R2": r2}

# [분류 성능지표] 혼동행렬에서 지표를 뽑아 딕셔너리로 반환
def get_cla_score(y_true, y_pred, y_prob):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()   # 혼동행렬 4칸 분해
    recall = recall_score(y_true, y_pred)         # 재현율(민감도): 실제 양성 중 맞춘 비율
    specificity = tn / (tn + fp)                  # 특이도: 실제 음성 중 맞춘 비율
    return {
        "Accuracy": accuracy_score(y_true, y_pred),    # 정확도
        "Recall": recall,                               # 재현율
        "Precision": precision_score(y_true, y_pred),   # 정밀도
        "Specificity": specificity,                     # 특이도
        "F1": f1_score(y_true, y_pred),                 # 정밀도·재현율 조화평균
        "G-Mean": np.sqrt(recall * specificity),        # 재현율·특이도 기하평균(불균형 대응)
        "AUC": roc_auc_score(y_true, y_prob),           # ROC 곡선 아래 면적
    }
```

```python
# 성능 평가 - 회귀
# 모델별 클래스 예측값과 양성 확률로 성능지표를 산출해 표로 정리 (MAE 기준 오름차순 정렬 (오차 작은 순))

reg_results = {}
for name in reg_models:
    reg_results[name] = get_reg_score(Y_reg_test_final, reg_preds[name])

reg_score_df = pd.DataFrame(reg_results).T.round(4).sort_values("MAE")
reg_score_df
```

```python
# 성능 평가 - 분류
# 모델별 클래스 예측값과 양성 확률로 성능지표를 산출해 표로 정리 (F1 기준 내림차순 정렬 (성능 좋은 순))

cla_results = {}
for name in cla_models:
    cla_results[name] = get_cla_score(Y_cla_test_final, cla_preds[name], cla_probs[name])

cla_score_df = pd.DataFrame(cla_results).T.round(4).sort_values("F1", ascending=False)
cla_score_df
```

```python
from sklearn.feature_selection import mutual_info_regression

# 연속형 X 변수만 사용 (타깃 y 제외)
X = Regression.drop(columns=["y"])
cols = X.columns.tolist()
n = len(cols)

# 변수쌍 MI 행렬 (대칭)
mi_matrix = np.zeros((n, n))

for i in range(n):
    for j in range(i + 1, n):
        mi = mutual_info_regression(
            X[[cols[i]]],           # feature
            X[cols[j]],             # target처럼 취급 (연속-연속 MI)
            random_state=42,
            n_neighbors=3,          # 샘플 수(~241)에 맞게 작게
        )[0]
        mi_matrix[i, j] = mi
        mi_matrix[j, i] = mi

# 대각선은 자기 자신과의 MI → NaN 처리
np.fill_diagonal(mi_matrix, np.nan)

mi_df = pd.DataFrame(mi_matrix, index=cols, columns=cols)

# Heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(
    mi_df,
    cmap="YlOrRd",
    square=True,
    linewidths=0.1,
    cbar_kws={"label": "Mutual Information"},
)
plt.title("Regression - Pairwise MI (연속형 변수)")
plt.tight_layout()
plt.show()
```

```python
# 상삼각만 추출해서 쌍 리스트 만들기
pairs = []
for i in range(n):
    for j in range(i + 1, n):
        pairs.append({
            "var1": cols[i],
            "var2": cols[j],
            "MI": mi_matrix[i, j],
            "Pearson": X[cols[i]].corr(X[cols[j]]),  # 비교용 (선형)
        })

mi_pairs_df = (
    pd.DataFrame(pairs)
    .sort_values("MI", ascending=False)
    .reset_index(drop=True)
)

print("=== MI 상위 20쌍 ===")
mi_pairs_df.head(20).round(4)
```

```python
mi_with_y = mutual_info_regression(
    X, Regression["y"], random_state=42, n_neighbors=3
)

mi_y_df = (
    pd.DataFrame({"feature": cols, "MI_with_y": mi_with_y})
    .sort_values("MI_with_y", ascending=False)
    .reset_index(drop=True)
)

plt.figure(figsize=(10, 8))
sns.barplot(data=mi_y_df.head(20), y="feature", x="MI_with_y", palette="viridis")
plt.title("Top 20 features by MI with y")
plt.tight_layout()
plt.show()

mi_y_df.head(20).round(4)
```

### 5-2) 전처리 심화

**[과제 4]** 지수는 수치형 변수들의 스케일이 제각각이라 모델 학습 전에 스케일링이 필요하다고 판단했습니다. 그런데 Min-Max와 Standard 중 어떤 걸 써야 할지, 변수마다 다르게 적용해야 하는 건 아닌지 확신이 서지 않습니다. AI와 어떤 스케일링이 적합할지 논의하고, 정한 방식을 적용해 보세요.

```python
import numpy as np
import pandas as pd

num_cols_reg = X_reg_train.select_dtypes(include=['int64', 'float64']).columns

def summarize_distribution(df, cols):
    rows = []
    for col in cols:
        x = df[col]
        q1, q3 = x.quantile(0.25), x.quantile(0.75)
        iqr = q3 - q1
        lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
        n_outliers = ((x < lower) | (x > upper)).sum()

        rows.append({
            "변수": col,
            "skew": x.skew(),                          # 0에 가까울수록 대칭 (정규분포에 가까움)
            "kurtosis": x.kurt(),                       # 0에 가까울수록 정규분포와 꼬리 두께 비슷
            "이상치_개수(IQR기준)": n_outliers,
            "이상치_비율(%)": round(n_outliers / len(x) * 100, 2),
        })
    return pd.DataFrame(rows).sort_values("이상치_비율(%)", ascending=False)

dist_summary = summarize_distribution(X_reg_train, num_cols_reg)
dist_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 이상치 비율 상위 6개 변수만 그려보기 (전체 58개는 너무 많음)
top_cols = dist_summary["변수"].head(6).tolist()

fig, axes = plt.subplots(2, 6, figsize=(20, 6))

for i, col in enumerate(top_cols):
    sns.boxplot(x=X_reg_train[col], ax=axes[0, i], color="skyblue")
    axes[0, i].set_title(f"{col} - Boxplot")

    sns.histplot(X_reg_train[col], kde=True, ax=axes[1, i], color="salmon")
    axes[1, i].set_title(f"{col} - Histogram")

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import shapiro

def normality_test(df, cols, alpha=0.05):
    rows = []
    for col in cols:
        stat, p = shapiro(df[col])
        rows.append({
            "변수": col,
            "p-value": p,
            "정규분포 여부": "정규분포 아님" if p < alpha else "정규분포에 가까움",
        })
    return pd.DataFrame(rows).sort_values("p-value")

normality_result = normality_test(X_reg_train, num_cols_reg)
normality_result
```

```python
result = dist_summary.merge(normality_result, on="변수")
result["정규분포 아님 개수"] = (result["정규분포 여부"] == "정규분포 아님").astype(int)

print("이상치 비율 5% 이상 변수 개수:", (result["이상치_비율(%)"] >= 5).sum())
print("정규분포 아님 변수 개수:", result["정규분포 아님 개수"].sum(), "/", len(result))

result.head(10)
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.linear_model import Ridge

# alpha 후보: 로그 스케일로 넓게 탐색 (0.001 ~ 1000)
param_grid = {"alpha": np.logspace(-3, 3, 20)}

kf = KFold(n_splits=5, shuffle=True, random_state=42)

grid_search = GridSearchCV(
    estimator=Ridge(),
    param_grid=param_grid,
    scoring="neg_mean_absolute_error",   # MAE 기준 (작을수록 좋음 → neg 처리)
    cv=kf,
    n_jobs=-1,
)

grid_search.fit(X_reg_train_final, Y_reg_train_final)

print("최적 alpha:", grid_search.best_params_)
print("최적 CV 성능 (MAE):", -grid_search.best_score_)
```
